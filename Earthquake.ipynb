```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression(max_iter=100000)
```

```python
lr.fit(X_train,y_train)
```

```python
pred_lr=lr.predict(X_test)
```

```python
print(classification_report(y_test,pred_lr))
```
